```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('sales_data.csv', delimiter =',', dtype = str, skip_header = 1)
days = data[:, 0]
items = data[:, 1]
sales = data [:, 2].astype(float)
print(data)
# print(items[:5])
# print(type(items[0]))
    
```

```python
each_day = np.unique(days)
day_totals = [np.sum(sales[days == d]) for d in each_day]
```

```python
plt.figure(figsize =(6,6))
plt.pie(day_totals, labels = each_day)
plt.title('Total Sales per Day')
plt.show()
```

```python
burger_sales = np.sum(data[1:,1].astype("int64"))
pizza_sales = np.sum(data[1:,2].astype("int64"))
coke_sales = np.sum(data[1:,3].astype("int64"))
fries_sales = np.sum(data[1:,4].astype("int64"))
cake_sales = np.sum(data[1:,5].astype("int64"))

print(f"burger_sales:", burger_sales )
print(f"pizza_sales:", pizza_sales )
print(f"coke_sales:", coke_sales )
print(f"fries_sales:", fries_sales )
print(f"cake_sales:", cake_sales )

```

```python
# Plotting
items = ['Burger', 'Pizza', 'Coke', 'Fries', 'Cake']
sales = [burger_sales, pizza_sales, coke_sales, fries_sales, cake_sales]

plt.figure(figsize=(8,5))
plt.bar(items, sales, color='mediumslateblue')
plt.title('Total Sales per Item')
plt.xlabel('Items')
plt.ylabel('Units Sold')
plt.grid(axis='y', linestyle='--', alpha=0.5)
for i, value in enumerate(sales):
    plt.text(i, value +5, str(value), ha='center', fontsize=10)
plt.show()
```

```python
items = ['Burger', 'Pizza', 'Coke', 'Fries', 'Cake']
sales = [burger_sales, pizza_sales, coke_sales, fries_sales, cake_sales]

max_index = np.argmax(sales)
min_index = np.argmin(sales)

highest_item = items[max_index]
lowest_item = items[min_index]
highest_sales = sales[max_index]
lowest_sales = sales[min_index]

print(f"Highest Selling Item: {highest_item} ({highest_sales})")
print(f"Lowest Selling Item: {lowest_item} ({lowest_sales})")


```

```python
extreme_items = [highest_item, lowest_item]
extreme_sales = [highest_sales, lowest_sales]
colors = ['darkorange', 'steelblue']

plt.figure(figsize=(6,4))
plt.bar(extreme_items, extreme_sales, color=colors)
plt.title('Highest vs Lowest Selling Items')
plt.xlabel('Items')
plt.ylabel('Units Sold')
plt.grid(axis='y', linestyle='--', alpha=0.4)



plt.tight_layout()
plt.show()
```
