```python
#highest sales and lowest sales

item_names =  working_data[0,1:]
numeric_data = working_data[1:,1:].astype(int)
total_sales= np.sum(numeric_data, axis = 0)
max_index = np.argmax(total_sales)
min_index = np.argmin(total_sales)
print("Total Sales Per Item:")
for item, total in zip(item_names, total_sales):
    print(f"{item}: {total}")

print("\n Highest Selling Item:")
print(f"{item_names[max_index]}: {total_sales[max_index]}")

print("\n Lowest Selling Item:")
print(f"{item_names[min_index]}: {total_sales[min_index]}")



```

```python
colors = ['skyblue'] * len(item_names)
colors[max_index] = 'green'   # Highest
colors[min_index] = 'red'     # Lowest

plt.figure(figsize=(10, 6))
bars = plt.bar(item_names, total_sales, color=colors)
plt.title("Total Sales Per Item")
plt.xlabel("Item")
plt.ylabel("Total Sales")
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10, str(height), ha='center', va='bottom')

# Highlight legend
plt.legend(handles=[
    plt.Rectangle((0,0),1,1,color='green', label='Highest'),
    plt.Rectangle((0,0),1,1,color='red', label='Lowest')
])
plt.tight_layout()
plt.show()

```
